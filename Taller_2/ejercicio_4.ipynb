```python
import numpy as np
import matplotlib.pyplot as plt

N = np.linspace(10, 100, 200)
a = 0
b = np.pi
```

```python
def MonteCarlo(funcion, N):
    x = np.random.uniform(a,b,int(N))
    return (b-a)*sum(funcion(x))/N
```

```python
def func_integrate(x):
    return np.exp(-x)*np.sin(x)
```

```python
Iteo = 0.5*(1+np.exp(-np.pi))
```

```python
Integral = np.array([])
for i in N:
    Integral = np.append(Integral, MonteCarlo(func_integrate, i))
```

```python
error_porcentual = np.array([])
for i in Integral:
    error_porcentual = np.append(error_porcentual, np.abs(1-(i/Iteo)))
```

```python
raiz_N = np.array([])
for i in N:
    raiz_N = np.append(raiz_N, (1/np.sqrt(i)))
```

```python
plt.scatter(N, error_porcentual, color = 'b')
plt.plot(N, raiz_N, color = 'r')
plt.xlabel('N')
plt.ylabel('Error porcentual')
plt.title('Error porcentual por un valor de N')
plt.show()
```
