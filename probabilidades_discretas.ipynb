```python
N = 100000
p = [0, 0.05, 0.1, 0.2, 0.3, 0.4]
P = np.linspace(0, 1, 200)
t = []
T = []
for j in range(len(P)):
    y = np.random.binomial(5, P[j], N)
    x = np.random.binomial(25, P[j], N)
    k = 0
    K = 0
    for i in y:
        if i >= 1:
            k += 1
    t.append(1-(k/N))
    for i in x:
        if i >= 5:
            K += 1
    T.append(1-(K/N))
```

```python
plt.plot(P, t, color = 'r', label = 'n = 5')
plt.plot(P, T, color = 'b', label = 'n = 25')
plt.axvline(x = 0.3, color = 'g')
plt.axvline(x = 0.1, color = 'c')
plt.legend(loc = 'upper right')
plt.show()
```
